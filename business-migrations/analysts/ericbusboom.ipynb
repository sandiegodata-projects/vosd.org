# California Business Migrations

This analysis uses an extract of the [National Establishment Time Series](http://exceptionalgrowth.org/our-databases.iegc) database to study business migrations into and out of California, with an emphasis on the San Diego metro region. 

Note that all of the records refer to Establishments, not Firms.  Refer to the [US Census website](https://ask.census.gov/faq.php?id=5000&faqId=487) for a formal definition of these terms. 

See the [manifest documentation](http://restricted.sandiegodata.org/manifests/134675d9-2845-458b-a153-3a3ab1d2dec7/documentation.html) for the data documentation.


Analysts, be sure to install or update sdipylib: `sudo pip install -U sdipylib`

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys 
# If either the import or the assert fails, you need to run: sudo pip install -U sdipylib
import sdipylib
assert sdipylib.__version__ >= 1005
```

```python
fn = sdipylib.url.cache_url('s3://restricted.sandiegodata.org/manifests/134675d9-2845-458b-a153-3a3ab1d2dec7/camigrations.csv')
print "Reading ",fn
df = pd.read_csv(fn)

# These assertions check values taken from the Excel Pivot Table, to check that we
# have the same data. 
assert df[(df.destcbsa == 31084) & (df.originstate != 'CA')].jobs.sum() == 78148
assert df[(df.destcbsa == 42044) & (df.origincbsa == 20940)].jobs.sum() == 68
assert df[(df.destcbsa == 31084)  & (df.origincbsa == 31084) ].estabs.sum() == 173234 
assert df[(df.destcbsa == 31084)  & (df.originstate == 'CA') ].estabs.sum() == 188257 
assert df[(df.origincbsa == 35084)  & (df.destcbsa == 33700 ) ].jobs.sum() == 5
```

```python
simple_cols = ['moveyear','industry','origin_metro_area','destination_metro_area','estabs','jobs'] # Columns for table views. 
```

```python
# Create an alternative dataframe 'Half DF' that just lists the number of jobs going into or out 
# of each CBSA
xdf = df[df.origincbsa != df.destcbsa ] # Start with only the records that moved CBSA

# The "out" half
odf = xdf[['moveyear','industry','originstate','origincbsa','origin_metro_area','jobs','estabs']]
odf.columns = ['year','industry','state','cbsa','metro','jobs','estabs']
odf['sig_jobs'] = -odf['jobs']
odf['sig_estabs'] = -odf['estabs']

# The "in" half
idf = xdf[['moveyear','industry','deststate','destcbsa','destination_metro_area','jobs','estabs']]
idf.columns = ['year','industry','state','cbsa','metro','jobs','estabs']
idf['sig_jobs'] = idf['jobs']
idf['sig_estabs'] = idf['estabs']

# Join into a single DataFrame
hdf = odf.append(idf, ignore_index = True)
```

```python
# Add a 'State_direction' column to make it easier to sum up the jobs into and out of the state. 
# dir == 0 means business stayed in state, 1 == moved in, -1 == moved out
df['state_dir'] = df.apply(lambda x: (int(x.originstate == 'CA') * -1 ) + (int(x.deststate == 'CA')), axis = 1 )
```

## Migrations Per Year, For CA and San Diego

First, let's look at the net number of jobs that entered and left California, by year. Positive numbers mean jobs or establishments moved in, negative means they moved out. 


```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (15,4))
hdf[hdf.state == 'CA'].sort('year').groupby('year').sum().sig_jobs.plot(ax=axes[0],title='California Job Migrations Per Year')
hdf[hdf.state == 'CA'].sort('year').groupby('year').sum().sig_estabs.plot(ax=axes[1],title='California Establishment Migrations Per Year')
```

And here are the cumulative values. They are relentlessly negative. 

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (15,4))
hdf[hdf.state == 'CA'].sort('year').groupby('year').sum().cumsum().sig_jobs.plot(ax=axes[0],title='California Cumulative Job Migrations Per Year')
hdf[hdf.state == 'CA'].sort('year').groupby('year').sum().cumsum().sig_estabs.plot(ax=axes[1],title='California Cumulative Establishment Migrations Per Year')
```

Next, let's look at the net number of jobs that entered and left San Diego metro, by year.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (15,4))
hdf[hdf.cbsa == 41740].sort('year').groupby('year').sum().sig_jobs.plot(ax=axes[0],title='San Diego Job Migrations Per Year')
hdf[hdf.cbsa == 41740].sort('year').groupby('year').sum().sig_estabs.plot(ax=axes[1],title='San Diego Establishment Migrations Per Year')
```

Unlike the state as a whole, the cumulative charts for San Diego county are almost entirely positive. 

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (15,4))
hdf[hdf.cbsa == 41740].sort('year').groupby('year').sum().cumsum().sig_jobs.plot(ax=axes[0],title='San Diego Cumulative Job Migrations Per Year')
hdf[hdf.cbsa == 41740].sort('year').groupby('year').sum().cumsum().sig_estabs.plot(ax=axes[1],title='San Diego Cumulative Establishment Migrations Per Year')
```

## Top Origin and Destination States From San Diego

These barcharts show the top states that are origins or destinations for jobs and establishments that arrived in or left the San Diego metro.


```python
arrived_ca = df[df.destcbsa == 41740 ].groupby('originstate')[['jobs','estabs']].sum()

fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (15,12))
plt.subplots_adjust(hspace = .5)
arrived_ca.sort('jobs', ascending=False)[1:20].jobs.plot(kind='bar', ax=axes[0], figsize = (15,5))
axes[0].set_title('Jobs arrived in San Diego Metro')
arrived_ca.sort('estabs', ascending=False)[1:20].estabs.plot(kind='bar', ax=axes[1], figsize = (15,5))
axes[1].set_title('Estabs arrived in San Diego Metro')
```

The large number of jobs from South Dakota is primarily one firm that moved in 1999. 


```python
(df[(df.destcbsa == 41740) & ( df.originstate == 'SD') ]
 [['moveyear','industry','origin_metro_area','destination_metro_area','estabs','jobs']]
 .sort('moveyear'))
```

However, the Texas moves are much more distributed, with a lot of smaller firms over many years. 

```python
(df[(df.destcbsa == 41740) & ( df.originstate == 'TX') ]
 [['moveyear','industry','origin_metro_area','destination_metro_area','estabs','jobs']]).groupby('moveyear').sum()
```

Here are the jobs and establishment counts of those that left the San Diego metro. 

```python
left_ca = df[df.origincbsa == 41740 ].groupby('deststate')[['jobs','estabs']].sum()

fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (15,12))
plt.subplots_adjust(hspace = .5)
left_ca.sort('jobs', ascending=False)[1:20].jobs.plot(kind='bar', ax=axes[0], figsize = (15,5))
axes[0].set_title('Jobs left the San Diego Metro')
left_ca.sort('estabs', ascending=False)[1:20].estabs.plot(kind='bar', ax=axes[1], figsize = (15,5))
axes[1].set_title('Estabs left the San Diego Metro')
pass
```

Finally, the net counts, where positive number indicate a net inflow. Here are the 20 most positive values. 

```python
net_ca = (arrived_ca - left_ca).dropna()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (15,12))
plt.subplots_adjust(hspace = .5)
net_ca.sort('jobs', ascending=False)[1:21].jobs.plot(kind='bar', ax=axes[0], figsize = (15,5))
axes[0].set_title('Net jobs for San Diego Metro')
net_ca.sort('estabs', ascending=False)[1:21].estabs.plot(kind='bar', ax=axes[1], figsize = (15,5))
axes[1].set_title('Net Estabs for San Diego Metro')
pass
```

And here are the 20 most negative. 

```python
net_ca = (arrived_ca - left_ca).dropna()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (15,12))
plt.subplots_adjust(hspace = .5)
net_ca.sort('jobs', ascending=True)[1:21].jobs.plot(kind='bar', ax=axes[0], figsize = (15,5))
axes[0].set_title('Net jobs for San Diego Metro')
net_ca.sort('estabs', ascending=True)[1:21].estabs.plot(kind='bar', ax=axes[1], figsize = (15,5))
axes[1].set_title('Net Estabs for San Diego Metro')
```
