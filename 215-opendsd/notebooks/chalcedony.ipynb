```python
import sqlite3
import json
```

```python
def gen(cur):
    for row in cur.execute('SELECT data FROM json'):
        d = json.loads(str(row[0]).decode('zlib'))
        
        if 'ProjectId' in d:
            if d['ProjectId'] == 397780:
                yield d
```
