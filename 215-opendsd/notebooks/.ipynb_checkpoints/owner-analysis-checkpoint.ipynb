```python
import pandas as pd
import numpy as np
```

```python
DIR = "/Users/damoncrockett/vosd.org/215-opendsd/tables/"
```

```python
fees = pd.read_csv(DIR+'fees_project_year.csv')
```

```python
parcels = pd.read_csv(DIR+'approvals_parcels_CPA.csv')
del parcels['Unnamed: 0']
```

```python
parcels = parcels[['ProjectId',
                   'StreetAddress',
                   'Latitude',
                   'Longitude',
                   'asr_land',
                   'asr_impr',
                   'asr_total',
                   'shape_area',
                   'own_name1',
                   'nucleus_zo',
                   'nucleus_us',
                   'unitqty',
                   'lat',
                   'lon',
                   'community_name']]
```

```python
parcels = parcels[parcels.lat.notnull()]
parcels.reset_index(drop=True,inplace=True)
```

```python
parcels.set_index('ProjectId',inplace=True)
```

```python
d = fees.join(parcels,on='ProjectId')
```

```python
np.mean(d.isnull())
```

```python
d = d[d.lat.notnull()]
```

## d is our data. The rows are invoices.

One issue: The parcel data is current. The invoices are not. So the current owner is associated with all the development activity since 2003, which may be wrong in some cases.

```python
d.info()
```

```python
tmp = d[['own_name1',
         'asr_total',
         'asr_land',
         'asr_impr',
         'TotalAmount']]
t = tmp.groupby('own_name1').agg(np.median)
t['count'] = tmp.groupby('own_name1').size()
t.sort('count',
       ascending=False)[:50].reset_index().sort('TotalAmount',
                                                ascending=False).reset_index(drop=True).to_csv(DIR+'owners.csv',
                                                                                               index=False)
```

```python
t.sort('count',
       ascending=False)[:50].reset_index().sort('TotalAmount',
                                                ascending=False).reset_index(drop=True)
```

```python
tmp = d[['own_name1',
         'asr_total',
         #'asr_land',
         'asr_impr',
         'TotalAmount']]
t = tmp.groupby('own_name1').agg({'asr_total':np.median,
                                  'asr_impr':np.median,
                                  'TotalAmount':np.sum})
t['count'] = tmp.groupby('own_name1').size()
t.sort('count',
       ascending=False)[:50].reset_index().sort('TotalAmount',
                                                ascending=False).reset_index(drop=True).to_csv(DIR+'owners_tot.csv',
                                                                                               index=False)
```
