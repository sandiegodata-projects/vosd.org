```python
import pandas as pd
import sqlite3
import json
```

```python
# create connection and cursor for database file 'approvals.db', already downloaded

con = sqlite3.connect('approvals.db')
cur = con.cursor()
```

```python
# generator to grab a collection of relevant variables from each record

def gen_approvals(cur):
    for row in cur.execute('SELECT data FROM json'):
        d = json.loads(str(row[0]).decode('zlib'))
        
        
        """
        Consider only those records that have Approval, Job, and
        Project keys. In order to get the variables we want, we
        need only these.
        """
        
        if 'Approval' and 'Job' and 'Project' in d:
            
            """
            Create a temporary dict by merging Approval, Job and 
            Project dicts. The purpose of this is to obtain a 
            flattened, unified dict that contains all the keys
            we want.
            """
            
            temp = dict(d['Approval'].items() + 
                       d['Job'].items() + 
                       d['Project'].items())
            
            # make a list of the keys we want
            kept_keys = ['ApprovalId',
                 'ProjectId',
                 'ApplicationDate',
                 'IssueDate',
                 'IssuedBy',
                 'Type',
                 'DevelopmentTitle',
                 'Latitude',
                 'Longitude',
                 'SortableStreetAddress']
            
            # dict comprehension to yield only the keys we want
            yield {key:temp[key] for key in kept_keys}
```

```python
df = pd.DataFrame(gen_approvals(cur))
```
