# Preliminary: Data Mining and Cleaning

```python
%matplotlib inline
import pandas as pd
import sqlite3
import json
import dateutil.parser
import datetime
import os
import gzip
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

### Download database file from the repository

```python
def download_ambry_db(url,name):  
    """Download an Ambry database file from the repository. """
    import gzip
    import sys
    
    if not os.path.exists(name+'.gz'):
        print 'Downloading:', url
        sys.stdout.flush()
        import urllib
        urllib.urlretrieve (url, name+'.gz')
    else:
        print 'Already downloaded:', url
        sys.stdout.flush()

    if not os.path.exists(name):
        print 'Extracting to:',name
        sys.stdout.flush()
        with open(name,'wb') as out_f, gzip.open(name+'.gz', 'rb') as in_f:
            out_f.write(in_f.read())   
    else:
        print 'Already extracted:', name
        sys.stdout.flush()
```

```python
download_ambry_db("http://s3.sandiegodata.org/library/sandiego.gov/opendsd-json-0.0.1/json/approvals.db", "approvals.db")  
```

### Create appropriate dataframe from database file

```python
# create connection and cursor for database file 'approvals.db'

con = sqlite3.connect('/home/damoncrockett/Desktop/approvals.db')
cur = con.cursor()
```

```python
#row = cur.execute('SELECT data FROM json').fetchone()
#d = json.loads(str(row[0]).decode('zlib'))
#d
```

```python
# generator to grab a collection of relevant variables from each record

def gen_approvals(cur):
    for row in cur.execute('SELECT data FROM json'):
        d = json.loads(str(row[0]).decode('zlib'))
        
        if 'Approval' and 'Job' and 'Project' in d:
            temp = dict(d['Approval'].items() + 
                       d['Job'].items() + 
                       d['Project'].items())
            
            kept_keys = ['ApprovalId',
                 'ProjectId',
                 'ApplicationDate',
                 'IssueDate',
                 'IssuedBy',
                 'Type',
                 'Latitude',
                 'Longitude',
                 'StreetAddress',
                 'PermitHolder',
                 'MapReference',
                 'Scope']
            
            yield {key:temp[key] for key in kept_keys}
```

```python
# create dataframe from generator
df = pd.DataFrame(gen_approvals(cur))
```

### Clean up dataframe

```python
# remove records without application dates
df = df[df.ApplicationDate.notnull()]
```

```python
# remove records without issue dates
df = df[df.IssueDate.notnull()]
```

```python
# parse ISO 8601 and remove time
df.ApplicationDate = df.ApplicationDate.apply(dateutil.parser.parse)
df.IssueDate = df.IssueDate.apply(dateutil.parser.parse)
df.ApplicationDate = df.ApplicationDate.apply(datetime.datetime.date)
df.IssueDate = df.IssueDate.apply(datetime.datetime.date)
```

```python
# compute time to issue
df['toIssue'] = df.IssueDate - df.ApplicationDate
```

```python
# eliminate rows with impossible toIssue values
df = df[df.toIssue >= 0]
```

```python
# convert toIssue column to an integer value by dividing by a timedelta of 1 day
df['DaystoIssue'] = df.toIssue / np.timedelta64(1, 'D')
```

```python
# eliminate toIssue column
del df['toIssue']
```

#### Answer: Yes, the relationship is statistically significant (p < 0.001), but the correlation is weak. The larger the project, the longer each individal approval takes. Though again, this correlation is pretty weak at r = 0.12.

## Case Studies: Efficient Pullers

As a kind of aside, I want to consider a few very efficient individuals to see whether they tend toward certain areas of San Diego, or certain developers or permit types, and whether this might at least partially explain their success. We already know that sheer number of permits pulled is not correlated with time to issue. So, it can't be that. But there is a quite significant effect of permit issuer on speed, and that demands some sort of explanation. If speed is not correlated with any of these other things, then we'll have to put it down to some intrinsic talent.
