### CU and NU permits

```python
tmp = df[(df.Type == 'Conditional Use Permit') | (df.Type == 'Neighborhood Use Permit')]
tmp = tmp[['Type','TimetoIssue','ApplicationYear']]
by_year = tmp.groupby(['Type','ApplicationYear'])
tmp = by_year.aggregate([np.median,
                   max,
                   min])
tmp['Count'] = by_year.size()
#tmp.to_csv('/home/damoncrockett/vosd.org/215-opendsd/CU_NU_time_year_count.csv')
tmp
```

```python
app_year = bau.ApplicationYear.apply(str)
app_month = bau.ApplicationMonth.apply(str)
app_month = app_month.map(lambda x: x.zfill(2))
```
